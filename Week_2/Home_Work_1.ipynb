```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import scale
```

```python
data = pd.read_table('wine.data', delimiter = ',', header = 0, index_col = None)
```

```python
data.head()
```

```python
target = data['Class']
train = data[['Alcohol','Malic_acid','Ash','Alcalinity_of_ash','Magnesium','Total_phenols','Flavanoids','Nonflavanoids_fenols','Proanthrocyanins','Color_intesity','Hue','OD280/OD315_of_diluted_wines','Proline']]
kf = KFold(n_splits=5, shuffle=True, random_state=42)
neigh = KNeighborsClassifier(n_neighbors=1)
a = np.mean(cross_val_score(neigh, train, target, cv=kf))
c = 1
for k in range(1, 51):
    neigh = KNeighborsClassifier(n_neighbors=k)
    b = np.mean(cross_val_score(neigh, train, target, cv=kf))
    if b > a:
        c = k
        a = b
file = open('1.txt', 'w')
file.write(str(c))
file.close()
file = open('2.txt', 'w')
file.write(str(a))
file.close()
```

```python
new_train = scale(train)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
neigh = KNeighborsClassifier(n_neighbors=1)
a = np.mean(cross_val_score(neigh, train, target, cv=kf))
c = 1
for k in range(1, 51):
    neigh = KNeighborsClassifier(n_neighbors=k)
    b = np.mean(cross_val_score(neigh, new_train, target, cv=kf))
    if b > a:
        c = k
        a = b
file = open('3.txt', 'w')
file.write(str(c))
file.close()
file = open('4.txt', 'w')
file.write(str(a))
file.close()
```
