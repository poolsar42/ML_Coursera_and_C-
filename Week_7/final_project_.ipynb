```python
#import data 
import pandas

features = pandas.read_csv('/Users/liliasergeenkova/Downloads/features.csv')

#delete features
y_data = features['radiant_win']
del features['duration']
del features['radiant_win']
del features['tower_status_radiant']
del features['tower_status_dire']
del features['barracks_status_radiant']
del features['barracks_status_dire']

x_data = features.iloc[:, 2:]
```

```python
# find the gaps
n = len(features['start_time'])
features_len = features.count()
gaps = []
m = len(features_len)
for i in range(0, m):
    if (features_len[i] < n):
        gaps.append(list(features)[i])
count_gaps = len(gaps)
```

```python
# fill the gaps
for i in range (len(gaps)):
    x_data.loc[:, gaps[i]].fillna(value=0,inplace=True) 
x_data_ = x_data.copy(deep=True)
```

```python
# GradientBoostingClassifier

from sklearn.model_selection import KFold 
from sklearn.ensemble import GradientBoostingClassifier
import time
import datetime
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit_transform(x_data)

kf = KFold(shuffle=True)

grid = {'n_estimators': [10, 20, 30, 40, 50, 60, 70]}
gbc = GradientBoostingClassifier(verbose=True) 
gs = GridSearchCV(gbc, grid, scoring='roc_auc', cv=kf)
gs.fit(x_data, y_data)
```

```python
best_p = gs.best_params_['n_estimators']
best_sc = gs.best_score_
print(best_p, best_sc)
```

```python
# time

import time
import datetime
from sklearn.model_selection import cross_val_score
start_time = datetime.datetime.now()

gbc = GradientBoostingClassifier(n_estimators = 30, verbose=True) 

score = cross_val_score(gbc, x_data, y_data, cv=kf, scoring='roc_auc')

time = datetime.datetime.now() - start_time
print(time, score.mean())
```

```python
# LogisticRegression

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
import numpy as np


grid = {'C': np.power(10.0, np.arange(-5, 6))}

lr = LogisticRegression()
gs = GridSearchCV(lr, grid, scoring='roc_auc', cv=kf)
gs.fit(x_data, y_data)

best_p = gs.best_params_['C']
best_sc = gs.best_score_
print(best_p, best_sc)
```

```python
# delete features 

del x_data['lobby_type']

names = []

for i in range(1, 6):
    s1 = 'r' + str(i) + '_hero'
    del x_data[s1]
    s2 = 'd' + str(i) + '_hero'
    del x_data[s2]
    names.append(s1)
    names.append(s2)

lr = LogisticRegression()
gs = GridSearchCV(lr, grid, scoring='roc_auc', cv=kf)
gs.fit(x_data, y_data)
best_p_del = gs.best_params_['C']
best_sc_del = gs.best_score_
print(best_p_del, best_sc_del)
```

```python
# pick
# add features from last block
import numpy as np

hero_c = [c for c in x_data_.columns if 'hero' in c]
all_heroes_id = np.unique(x_data_[hero_c])
wb = {}
for id in all_heroes_id:
    r = [(x_data_['r%d_hero' % n] == id) + 0 for n in range(1, 6)]
    d = [(x_data_['d%d_hero' % n] == id) + 0 for n in range(1, 6)]
    wb['hero%s' % id] = sum(r) - sum(d)
X_pick = x_data_.assign(**wb)
del X_pick['lobby_type']
X_pick
```

```python
# LogisticRegression for new features

lr = LogisticRegression()
gs = GridSearchCV(lr, grid, scoring='roc_auc', cv=kf)
gs.fit(X_pick, y_data)

best_p_pick = gs.best_params_['C']
best_sc_pick = gs.best_score_
print(best_p_pick, best_sc_pick)
```
