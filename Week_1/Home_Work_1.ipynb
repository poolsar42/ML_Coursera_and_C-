```python
import pandas as pd
```

```python
data = pd.read_csv('titanic.csv', index_col = 'PassengerId')
```

```python
woman = 0
man = 0
for sex in data['Sex']:
    if sex == 'female':
        woman += 1
    if sex == 'male':
        man += 1
file = open('1.txt', 'w')
file.write(str(man) + ' ')
file.write(str(woman))
file.close()
```

```python
#print (data['Survived'].describe())
name = data['Survived']
file = open('2txt', 'w')
dex = name.sum() / 891 * 100
file.write(str("%.2f" %dex))
file.close()
```

```python
k = 0
name = data['Pclass']
for i in name:
    if i == 1:
        k += 1
k = k / 891 * 100
file = open('3txt', 'w')
file.write(str("%.2f" %k))
file.close()
```

```python
print(data['Age'].describe())
file = open('4.txt', 'w')
file.write(str(data['Age'].mean()) + ' ')
file.write(str(data['Age'].median()))
file.close()
```

```python
data1 = data['SibSp']
data2 = data['Parch']
correlation = data1.corr(data2, method = 'pearson')
```

```python
print(correlation)
```

```python
file = open('5.txt', 'w')
file.write(str(correlation))
file.close()
```

```python
data3 = data['Name']
names = []
sex = []
words = []
for x in data3:
    names.append(x.split())
    sex.append(names[len(names)-1][1])
    for i in names[len(names) - 1]:
        words.append(i)
get = set(words)
prod = dict.fromkeys(get, 0)
for i in words:
    prod[i] += 1
sorted_names = sorted(prod.items(), key = lambda x: x[1], reverse = True)
sorted_names
```
